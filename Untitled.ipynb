```python
import os, sys, email
import numpy as np 
import pandas as pd
# Plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set_style('whitegrid')
#import plotly
#plotly.offline.init_notebook_mode()
#import plotly.graph_objs as go
import wordcloud

# Network analysis
import networkx as nx
# NLP
from nltk.tokenize.regexp import RegexpTokenizer
```

```python
def traverseDirectory(parentDirectory):
    # The top argument for walk
    topdir = parentDirectory
 
    # The extension to search for
    exten = '.txt'
    folderAnalysis=[]
    startinglevel = topdir.count(os.sep) 
    for dirpath, dirnames, files in os.walk(topdir):
        level = dirpath.count(os.sep) - startinglevel
        if(level==1):
            folderAnalysis.append(dirpath)
    return folderAnalysis
```

```python
def traverseFiles(directory):
    # The top argument for walk
    topdir = directory
 
    # The extension to search for
    exten = '.txt'
    filesAnalysis=[]
    # startinglevel = topdir.count(os.sep) 
    for dirpath, dirnames, files in os.walk(topdir):
        for name in files:
           filesAnalysis.append(os.path.join(dirpath,name))
    return filesAnalysis
```

```python
## Helper functions
def get_text_from_email(msg):
    '''To get the content from email objects'''
    parts = []
    for part in msg.walk():
        if part.get_content_type() == 'text/plain':
            parts.append( part.get_payload() )
    return ''.join(parts)

```

```python
def split_email_addresses(line):
    '''To separate multiple email addresses'''
    if line:
        addrs = line.split(',')
        addrs = frozenset(map(lambda x: x.strip(), addrs))
    else:
        addrs = None
    return addrs
```

```python
def prepareFrame(directoryList):
    filenames=[]
    df = pd.DataFrame()
    for folderName in directoryList:
        fileNames=traverseFiles(folderName)
        for file in fileNames:
            if '.DS_Store' not in file:
                #print(file)
                with open(file) as a_file:
                     df = df.append({'name': os.path.basename(folderName), 'file_name': file, 'content': a_file.read()}, ignore_index=True)
    return df
```

```python
initialList=['./maildir/allen-p','./maildir/arnold-j']
fd=prepareFrame(initialList)
```

```python
print(fd['content'][1])
```

```python
# Parse the emails into a list email objects
messages = list(map(email.message_from_string, fd['content']))
fd.drop('content', axis=1, inplace=True)
# Get fields from parsed email objects
keys = messages[0].keys()
for key in keys:
    fd[key] = [doc[key] for doc in messages]
# Parse content from emails
fd['body'] = list(map(get_text_from_email, messages))
# Split multiple email addresses
fd['From'] = fd['From'].map(split_email_addresses)
fd['To'] = fd['To'].map(split_email_addresses)

# Extract the root of 'file' as 'user'
#fd['user'] = fd['file_name'].map(lambda x:x.split('/')[0])
del messages

fd.head()
```

```python
print('shape of the dataframe:', fd.shape)
# Find number of unique values in each columns
for col in fd.columns:
    print(col, fd[col].nunique())
```

```python
# Set index and drop columns with two few values
fd = fd.set_index('Message-ID')\
    .drop(['file_name', 'Mime-Version', 'Content-Type', 'Content-Transfer-Encoding'], axis=1)
# Parse datetime
fd['Date'] = pd.to_datetime(fd['Date'], infer_datetime_format=True)
fd.dtypes
```

```python
# Count words in Subjects and content
tokenizer = RegexpTokenizer(r'(?u)\b\w\w+\b')
fd['subject_wc'] = fd['Subject'].map(lambda x: len(tokenizer.tokenize(x)))
fd['content_wc'] = fd['body'].map(lambda x: len(tokenizer.tokenize(x)))
```

```python

grouped_by_people = fd.groupby('name').agg({
        'body': 'count', 
        'subject_wc': 'mean',
        'content_wc': 'mean',
    })
grouped_by_people.rename(columns={'body': 'N emails', 
                                  'subject_wc': 'Subject word count', 
                                  'content_wc': 'Content word count'}, inplace=True)
grouped_by_people.sort('N emails', ascending=False).head()
```

```python
sub_df = fd[['From', 'To', 'Date']].dropna()
print(sub_df.shape)
# drop emails sending to multiple addresses
sub_df = sub_df.loc[sub_df['To'].map(len) == 1]
print(sub_df.shape)
```

```python

sub_df = sub_df.groupby(['From', 'To']).count().reset_index()
# Unpack frozensets
sub_df['From'] = sub_df['From'].map(lambda x: next(iter(x)))
sub_df['To'] = sub_df['To'].map(lambda x: next(iter(x)))
# rename column
sub_df.rename(columns={'Date': 'count'}, inplace=True)
sub_df.sort('count', ascending=False).head(10)
```

```python
# Make a network of email sender and receipients
G = nx.from_pandas_dataframe(sub_df, 'From', 'To', edge_attr='count', create_using=nx.DiGraph())
print('Number of nodes: %d, Number of edges: %d' % (G.number_of_nodes(), G.number_of_edges()))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 8))
ax1.hist(list(G.in_degree(weight='count').values()), log=True, bins=20)
ax1.set_xlabel('In-degrees', fontsize=18)

ax2.hist(list(G.out_degree(weight='count').values()), log=True, bins=20)
ax2.set_xlabel('Out-degrees', fontsize=18)
```

```python

n_nodes_in_cc = []
for nodes in nx.connected_components(G.to_undirected()):
    n_nodes_in_cc.append(len(nodes))

plt.hist(n_nodes_in_cc, bins=20, log=True)
plt.xlabel('# Nodes in connected components', fontsize=18)
plt.ylim([.1,1e4])
```
